## 준비 실습. Image 살펴보기

```python
## 1. package 가져오기
from PIL import Image
import numpy as np

print(Image.__version__)
```

```python
## 2. 이미지 가져오기
image_9 = Image.open('01image.png') # 이미지 객체 상태
print(image_9)
```

```python
## 3. 이미지 array로 변환하기

array_9 = np.array(image_9) # 이미지 객체를 사용하기 위해서 numpy 배열로 변경해준다.
print(array_9) # 자세히 보면 9라는 숫자가 보인다. 
```

```python
## 4. 이미지 살펴보기
print(array_9.shape) 
# 가로 28, 세로 28 채널은 1인데 흑백이라 채널이 1인 경우 생략이 된다.
# 이렇게 되면 keras에서 학습이 불가능하다.
# Dense layer input -> 1차원
# Convolution layer -> 3차원으로 둘 다 불가능하니 변경해주어야한다.
```

```python
## 5. 이미지 3차원 reshape : 이미지 픽셀의 값이나 개수는 변화가 없다. 데이터의 쌓는 방식이 변경된다.
array_9_3d = array_9.reshape(28,28,1)
# 원래 픽셀의 개수 = 28*28 = 784개
# 3차원일 때, 픽셀의 개수 = 28*28*1 = 784개
# 숫자값을 넣어서 변경할 때는 픽셀의 값을 고려해서 넣어주어야한다. 
```

```python
# 6. 이미지 1차원 reshape
array_9_1d = array_9.reshape(784) # 784 혹은 28*28로도 가능하다.
print(array_9_1d)
```

```python
## 이미지 시각화
import matplotlib.pyplot as plt

# 라이브러리가 이미지를 불러올 때에 RGB순서가 변경되서 이상하게 나오게 된다.
# 1차원은 더이상 이미지가 아니라서 출력이 안된다.
plt.imshow(array_9)
```

```python
## 이미지 시각화(흑백)

# cmap=plt.cm.gray를 추가해서 흑백으로 가능하다.
plt.imshow(array_9, cmap=plt.cm.gray)
```

```python
## 7. 이미지 resize : 픽셀을 쪼개는 작업

import cv2

# 위의 불러온 방식과 다르게 불러올 수 있다.
image_9_Cv = cv2.imread('01image.png')

image_9_Cv.resize(10,10) # 원하는 사이즈를 지정해둔다.

print(image_9_Cv.shape) # 이런 작업과 reshape랑은 다르다.
```
