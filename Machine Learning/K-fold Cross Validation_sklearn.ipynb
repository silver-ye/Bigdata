```python
# 1. 패키지 호출

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# 2. load data set

data = load_iris()

X = data.data #feature
y = data.target #label

print(X)
print(y)
```

```python
# 3. Train/Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 320)

# 4. Train
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# 5. Predict
y_pred = model.predict(X_test)

# 6. Evaluate
print(accuracy_score(y_pred, y_test)*100)
```

```python
# random_state를 못믿어서 사용하는 crocss validation
# Cross Validation

from sklearn.model_selection import KFold, cross_val_score

# random_state가 들어가지만 5번 진행해서 신뢰도를 높이는 것
data_kf = KFold(n_splits = 10, shuffle = True, random_state = 9)

score = cross_val_score(model, X, y, cv = data_kf, scoring = 'accuracy')

print('10 fold acc :', score)
print('Mean Acc :', score.mean())
```
