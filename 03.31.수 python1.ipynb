```python
a = int(input('첫번째 숫자를 입력해주세요: '))
b = int(input('두번째 숫자를 입력해주세요: '))
c = int(input('세번째 숫자를 입력해주세요: '))

if a < b and a < c:
    if b < c:
        print('큰 수는 ' + str(c) +'이고 ' + '작은 수는 ' + str(a) + '이다.')
    if b > c:
        print('큰 수는 ' + str(b) +'이고 ' + '작은 수는 ' + str(a) + '이다.')
    
if b < a and b < c:
    if a < c:
        print('큰 수는 ' + str(c) +'이고 ' + '작은 수는 ' + str(b) + '이다.')
    if a > c:
        print('큰 수는 ' + str(a) +'이고 ' + '작은 수는 ' + str(b) + '이다.')
    
if c < b and c < a:
    if b < a:
        print('큰 수는 ' + str(a) +'이고 ' + '작은 수는 ' + str(c) + '이다.')
    if b > a:
        print('큰 수는 ' + str(b) +'이고 ' + '작은 수는 ' + str(c) + '이다.')
```

```python
print('숫자를 입력하세요')
cnt = 0
num_sum = 0
num_max = 0

while cnt < 5:
    num = int(input())
    cnt = cnt + 1
    num_sum = num + num_sum
    if num > num_max:
        num_max = num


print('입력 끝')
print('입력받은 숫자들의 합은 ' + str(num_sum))
print('입력받은 숫자들의 평균은 ' + str(num_sum / 5))

    
```
