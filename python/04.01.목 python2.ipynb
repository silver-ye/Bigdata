## - 숫자 입력 받은 후 합, 평균, 최대값, 최솟값 출력 프로그램(강사님)

```python
#순서가 중요하다.

print('숫자를 입력하세요')

cnt = 0
sumA = 0
minV = 999
maxV = -999

txt = input()

while txt != '입력 끝' :
    
    txt = int(txt)
    sumA = sumA + txt
    cnt = cnt + 1
    
    if minV > txt :
        minV = txt
    if maxV < txt :
        maxV = txt
    
    txt = input()

print('총합: ', sumA)
print('평균: ', sumA / cnt)
print('최대: ', maxV)
print('최소: ', minV)
```

## - 최대값 최소값 찾기
- 길이가 20인 리스트에 무작위로 1~100사이의 숫자를 생성하여 저장하기
- 가장 큰 수와 가장 작은 수를 찾아서 출력
- 실제로 가장 큰 수와 가작 작은수를 확인
- while 과 for문으로 써보기

```python
#while문을 이용한 최대,최소값 찾기
import random

list_num = []
cnt = 0
max_num = 0
min_num = None

while cnt < 20:
    list_num.append(random.randint(1,100))
    if list_num[cnt] > max_num:
        max_num = list_num[cnt]
        if min_num == None:
            min_num = list_num[cnt]
    elif list_num[cnt] < min_num:
        min_num = list_num[cnt]
    cnt = cnt + 1

print('가장 큰 값' , max_num)
print('가장 작은 값', min_num)
print(list_num)
```

```python
#for문을 이용한 최대,최소값 찾기
import random

list_num = []
cnt = 0

while cnt < 20: #길이가 20인거를 카운팅한다.
    list_num.append(random.randint(1,100))
    cnt = cnt + 1
    
max_num = list_num[0] #첫번째 숫자를 넣어서 비교하기
max_num = list_num[0]

for num in list_num: #list안에 숫자들을 돌려서 num에 넣어서 비교하기
    if max_num < num:
        max_num = num
    if min_num > num:
        min_num = num

    

print('가장 큰 값' , max_num)
print('가장 작은 값', min_num)
print(list_num)
```

## - 알파벳 갯수를 구하고 출력하기

```python
#알파벳 개수를 출력하기
al = 'abcdefghijklmnopqrstuvwyz'
txt = '''Anaconda solutions are serious technology for real data science and ML applications. Anaconda is versatile you'll be ready to solve problems you don't even know you have yet.'''
txt = txt.lower()
for txts in al: #알파벳을 txts에 넣어서 돌리기
    cnt = 0
    for pattern in txt: #pattern에 txt를 넣어서 돌리기
        if txts == pattern: #알파벳을 넣은 txts와 txt가 들어가 pattern에서 동일한게 있다면
            cnt = cnt + 1 #동일한거를 카운트한다
    print(txts, '=', cnt, '개') #출력하기
        
```

## - While문으로 작성한 구구단 3단을 for문으로 변경하여 완성해보기

```python
# 구구단 3단을 while에서 for문으로 변경해보기
cnt = 1
while cnt < 10 :
    print('3 x', cnt, '=', 3*cnt)
    cnt = cnt + 1
    
print()    


for i in range(1,10):
    print('3 x', i, '=', 3*i)
```
