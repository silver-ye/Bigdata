## - 알고리즘

- 알고리즘이 의미: 문제를 해결하기 위한 정량적 단계적 절차(방법)

- 컴퓨터 알고리즘을 설명하기 위한 4단계
    • 문제 정의 (Problem definition) ex)배고프니 라면을 끓여먹겠다고 정의
    • 알고리즘 설명 (Algorithm description) ex)라면을 어떻게 끓여먹을지 
    • 정확성 증명 (Correctness proof) ex)라면을 먹음으로써 배고픔을 해소
    • 성능 분석 (Performance analysis) ex)라면을 끓이기위해서 비용등을 분석해서 좋은지 나쁜지 

- 컴퓨터 알고리즘의 수행시간의 분석은 어떻게 해야 하는가?
    • 조건이 동일한 특정 기계에서 모든 알고리즘의 수행시간을 측정해야하는데 현실적으로 불가능하다.
    •  그러므로 수행연산의 횟수를 비교하는 방식으로 성능을 분석한다.



## - 선택 정렬(Selection Sort)

- 선택 정렬(Selection sort)
    • 선택하여 정렬하는 알고리즘
    • 최소값 선택 정렬(오름차순)과 최대값 선택 정렬(내림차순), 보통 최솟값으로 정렬함.

```python
#수도코드를 참고하여 작성하기
###############################################
#for i <- 0 to n - 2
#min <- i
#for j <- i + 1 to n - 1
#    if A[j] < A[min] then
#        min <- j
#end for
#if min != i then interchange A[i] and A[min]
#end for
##############################################
A = ['7','1','9','5','3','8','4','6','4','2']
e = 0
n = len(A) #변수의 길이를 설정해서 인덱스값으로 생각하기

for i in range(0,n-1): #i와 j는 인덱스 값, len(A)는 10이고 인덱스값은 9니깐 n-1
    minA = i #minA에 인덱스값을 넣어준다.
    for j in range(i+1, n): #두 수를 비교하기
        if A[j] < A[minA]: #앞의 수 J와 minA의 값을 비교한다. 최솟값이니 앞의 수가 더 크다면 minA의 최솟값을 넣어서 저장하면서 돌아간다.
            minA = j
    if minA != i: #현재 바꾸고자하는 위치의 숫자와 제일 작은 값의 값이 같지 않다면
        e = A[i] #e라는 변수에 A[i]의 값을 저장해놓고
        A[i] = A[minA] #작은 값과 현재 위치의 값의 위치를 변경해준다.
        A[minA] = e 
        # 쉽게 생각하자면, 물이 담긴 컵과 오렌지쥬스가 담긴 컵에 내용물을 서로 바꾸고 싶어서 새로운 컵을 가져와서
        # 새로운 컵에 오렌지 쥬스를 넣고 물을 오렌즈쥬스컵에 넣고 새로운 컵에 있는 오렌즈쥬스는 물컵에 담으면 쉽게 바꿀 수 있다.
    
    print(i+1, '번째 ', list(A))

```

#Insertion Sort(A)
print('<삽입정렬>')
A = [5,1,6,9,2,7,0,3,8,4]
a = len(A)

for j in range(1,a):#1부터인 이유는 key값은 처음부터가 아닌 두번째값부터이고 두번째 값의 인덱스값은 1이다.
    key = A[j]
    i = j - 1 #i는 key값보다 앞에 있는 값을 비교해야해서 - 1을 해서 지정해준다.
    while i >= 0 and A[i] > key: #앞의 비교할 값(i)이 있으며, 현재값(key)이 작으면
        A[i + 1] = A[i] #현재값(key)은 앞으로가고 비교된 값(i)은 뒤로 간다.
        i = i - 1 #현재 위치보다 앞의 값을 다시 비교해주기 위해서 -1을 해서 앞의 값들과도 비교해준다.
    A[i+1] = key #앞의 비교할 값(i)이 더이상 없고, 현재값(key)이 더 크다면 그대로 위치한다.
    print(j, "번째 ", list(A))
