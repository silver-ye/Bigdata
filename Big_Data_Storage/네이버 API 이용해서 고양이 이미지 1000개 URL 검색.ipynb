```python
# 고양이와 강아지 이미지를 검색한다.
# keyword : 검색어, display : 검색 결과 갯수, start : 시작 인덱스
def get_api_result(keyword, display, start):
    # 네이버 이미지 검색
    url = "https://openapi.naver.com/v1/search/image?query="\
    + keyword + "&display="+ str(display) + "&start=" + str(start)
    print("url=", url)
    # 검색 결과를 result에 저장한다.
    result = requests.get(url,headers=api_id_pw)
    # 이미지 검색 결과를 리턴해준다.
    return result.json()
```

```python
# 한번에 100개씩이니 total_page를 100개로하면 만개
def call_and_print(keyword, total_page=10):
    link_list = []
    # 1에서 부터 total_page + 1미만 만큼을 반복한다.
    for page in range(1, total_page + 1):
        # 한 페이지당 100개를 가져온다.
        display = 100
        # 100개씩 검색하기떄문에 1씩 늘려서 검색해야한다
        start = ((page-1)*display)+1
        #print('page =', page)
        print('start =', start)
        # 검색해가지로 리턴을 받는 변수를 생성해준다.
        json_obj = get_api_result(keyword, display, start)
        # json_obj에 items의 link만 필요하다.
        for item in json_obj['items']:
            link_list.append(item['link'])
        #print('=' * 30)
    return link_list
```

```python
# link1은 위에서 지정한 고양이 이미지 링크
for url in link1:
    try: 
        # 이미지 내용을 다운받아서 image에 저장
        image = urllib.request.urlopen(url).read()
        # url을 split('/')을 통해서 나눠서 마지막에 위치한 이미지파일명을 받는다.
        image_name = url.split('/')[-1]
        
        # image_name에서 ?가 있다면 인덱스를 리턴하고, 없다면 -1을 리턴을 이용하여
        if image_name.find('?') != -1:
            # ?가 있다면 ?을 기준으로 나누고 인덱스[0]이 image_name이 된다.
            image_name = image_name.split('?')[0]

        # 이미지에 확장자가 없는 경우, 이미지의 기본 타입을 준다.
        image_type  ='jpg'

        # image_name에 .(점)이 위치하고 있다면 확장자를 나눠줄 수 있다.
        if image_name.find('.') != -1:
            # image_name에서 split('.')을 통해서 확장자를 나누고 마지막에 위치한 확장자 파일은 받는다
            image_type = image_name.split('.')[-1]

        # image의 확장자를 content_type 변수에 넣어주기
        content_type = 'image/{}'.format(image_type)

        # 이미지 파일의 정보를 변수에 저장한다.
        grid_in = bucket.open_upload_stream(
            # 이미지 파일의 이름과 확장자를 넣어준다.
            image_name,
            metadata = {'contentType':content_type, 'type':'cat'}
        )

        # 이미지의 내용을 저장하기
        grid_in.write(image)

        # 이미지 저장 종료 알려주기
        grid_in.close()
        print('image_type:', image_type, ':content_type:', content_type,
             ':image_name:', image_name)
    
    except:
        print('에러 발생')
```
