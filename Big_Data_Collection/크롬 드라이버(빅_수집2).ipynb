```python
# as로 별칭해주기
from selenium import webdriver as wd
```

```python
# 별칭 지정한거롤 사용
# 나온 크롬창은 끄면 안된다.
driver = wd.Chrome(executable_path = 'chromedriver.exe')
```

```python
# driver.get으로 naver로 보내기
driver.get('http://naver.com')
```

```python
# 검색어 입력하기
# driver.find_element_by_id는 시킨다.
# sent.keys()는 대신 입력하게한다.
driver.find_element_by_id('query').send_keys('윤여정')
```

```python
# 검색버튼을 검사로 찾고 그 버튼을 누르게 한다.
# 검색버튼: search_btn
driver.find_element_by_id('search_btn').click()
```

```python
# 페이지가 넘어가기전에 대기하는걸 load라고 한다.
# 페이지가 나오기전까지 10초를 기다린다.
# 10초 이전에 나오면 그냥 페이지가 나오고 무시된다.
driver.implicitly_wait(10)
```

```python
driver.implicitly_wait(10)
```

```python
# 페이지 넘기기
# 페이지를 넘기는 자바 스크립트 함수를 이용하기
# 현재 페이지가 7페이지까지니깐 7까지 돌린다.
import time
for page in range(1, 8):
    driver.execute_script(
        "searchModule.SetCategoryList({},'')".format(page)
    )
    
    #10초 멈추가하기(time.sleep()도 가능하지만, 무조건 그만큼 멈춘다.)
    time.sleep(3)
    
    # 빅데이터에서 문자열 + 하는건 파이썬 내부에서 시간이 오래 걸리기에
    # format을 쓰는 것이 더 낫다. 현재는 7개까지니 사용한다.
    print(str(page) + '번째 페이지 가요')
    
    boxItems = driver.find_elements_by_css_selector(
        ".panelZone>.oTravelBox>.boxList>li")
    
    # 가격인 proTit이 너무 많아서 h5인걸로 지정해준다.
    for li in boxItems:
        # h5.proTit에서 h5는 태그 .proTit은 클래스 명을 의미한다.
        #print("li.find_element_by_css_selector('h5.proTit')=",li.file_element_by_css_selector('h5.proTit')
        print('제목', li.find_element_by_css_selector('h5.proTit').text)      
        print('가격', li.find_element_by_css_selector('.proPrice').text.split('원')[0])
        print('=' * 100)
```

```python
import time
for page in range(1, 8):
    driver.execute_script(
        "searchModule.SetCategoryList({},'')".format(page)
    )
    
    
    time.sleep(3)
    print("{}번째 페이지로 이동".format(page))
    boxItems = driver.find_elements_by_css_selector(
        ".panelZone>.oTravelBox>.boxList>li")

    for li in boxItems:
        print('제목', li.find_element_by_css_selector('h5.proTit').text)      
        print('가격', li.find_element_by_css_selector('.proPrice').text.split('원')[0])
    print('=' * 100)
```
