```python
# 현재 딕셔너리로 이루어져있으며, items안에 리스트로 위치하고 있다.
result.json()['items']
```

```python
# for문 돌려서 리스트 하나씩 출력하기
# items에서 link만 필요하다.
# 10개만 출력된다.
for item in result.json()['items']:
    print('=' * 30)
    print("item['link']")
    print('=' * 30)
    print(item['link'])
    print('=' * 30)
```

```python
# 한번에 최대 100개까지만 가능하다.
# start=1 : 1번째 이미지부터, display=100 : 100번째 이미지까지 
url = 'https://openapi.naver.com/v1/search/image?query=고양이&display=100&start=1'
result = requests.get(url, headers = api_id_pw)
for item in result.json()['items']:
    print("item['link'] = ", item['link'])
    print('=' * 30)
```

```python
# 101번째 고양이부터 100개를 가져온다
url = 'https://openapi.naver.com/v1/search/image?query=고양이&display=100&start=101'
result = requests.get(url, headers = api_id_pw)
for item in result.json()['items']:
    print("item['link'] = ", item['link'])
    print('=' * 30)
```

```python
# 고양이와 강아지 이미지를 검색한다.
# keyword : 검색어, display : 검색 결과 갯수, start : 시작 인덱스
def get_api_result(keyword, display, start):
    # 네이버 이미지 검색
    url = "https://openapi.naver.com/v1/search/image?query="\
    + keyword + "&display="+ str(display) + "&start" + str(start)
    print("url=",url)
    # 검색 결과를 result에 저장한다.
    result = requests.get(url,headers=api_id_pw)
    # 이미지 검색 결과를 리턴해준다.
    return result.json()
```

```python
get_api_result('강아지', 100, 901)
```

```python
# 한번에 100개씩이니 total_page를 100개로하면 만개
def call_and_print(keyword, total_page=10):
    # 1에서 부터 total_page + 1미만 만큼을 반복한다.
    for page in range(1, total_page + 1):
        # 한 페이지당 100개를 가져온다.
        display = 100
        # 100개씩 검색하기떄문에 1씩 늘려서 검색해야한다
        start = ((page-1)*display)+1
        print('page =', page)
        print('start =', start)
        # 검색해가지로 리턴을 받는 변수를 생성해준다.
        json_obj = get_api_result(keyword, display, start)
        # json_obj에 items의 link만 필요하다.
        for item in json_obj['items']:
            print("item['link'] =", item['link'])
        print('=' * 30)
```

```python
# keyword를 강아지로 설정
call_and_print('강아지')
```

```python
# keyword를 고양이로 설정
call_and_print('고양이')
```
