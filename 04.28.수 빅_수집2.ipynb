```python
# 웹 페이지의 HTML을 가져오는 모듈(페이지 내용을 가져옴)
import requests
```

```python
# read와 같이 url의 페이지를 바로 가져온다.
# 가져오기 성공: 200, 가져오기 실패: 404(URL오타), 500(서버 오류)
# 위의 값을 리턴한다. 
# read해서 가져오는것보다 한번에 requests로 가져오는게 더 편하다.
requests.get('http://www.weather.go.kr/weather/observation/currentweather.jsp')
```

```python
# 변수에 담아주기
# response에는 2개가 저장된다
# (리턴한 값, content 속성(안에 가져온 html이 들어있음))
# response를 실행해서 리턴한 값을 통해서 성공했는지 확인이 가능하다.
response = requests.get('http://www.weather.go.kr/weather/observation/currentweather.jsp')
response
```

```python
# response.content로 html을 불러올 수 있다.
# 원래 웹에서는 한글이 숫자로 되어있는데 이것을 encoding이라고 한다.
response.content
```

```python
# response.text로도 불러오기가 가능하다.
# content는 encoding으로 나오지만 text로 불러오면 숫자가 아니라 한글로 변한다.
# 이를 decoding이라고 한다. 
# 즉, content는 encoding, text는 decoding되어서 나온다.
response.text
```

```python
# BeautifulSoup 한글이 숫자든 한글로 나오든 상관이 없다.
from bs4 import BeautifulSoup
```

```python
# BeautifulSoup으로 content사용
BeautifulSoup(response.content, 'html.parser')
```

```python
# BeautifulSoup으로 text사용
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# 데이터 크롤링 ppt 45p
# 기상청 지점 크롤링
# 하나니깐 find를 사용
table = soup.find('table',{'class': 'table_develop3'})
```

```python
for tr in table.find_all('tr'):
    print('tr = ', tr)
    print('=' *  100)
# 출력된 결과를 통해서 tr안에 td태그가 있음을 확인할 수 있다.
    
```

```python
# tr안에 있는 td를 찾아보기
for tr in table.find_all('tr'):
    # tr태그안에 있는 td를 찾기
    tds = tr.find_all('td')
    print('tds = ', tds)
    print('*' * 100)
    print('len(tds) = ', len(tds))  
    print('=' *  100)
# td중에서 인덱스[0]: 동네 이름, 인덱스[5]: 온도, 인덱스[10]: 습도
```

```python
# 인덱스에 아무것도 들어있지않은것들도 있다는 것을 확인했으니 if문으로 길이가 0이상인것을 지정해 리스트가 들어있는 것들만 찾는다.
for tr in table.find_all('tr'):
    tds = tr.find_all('td')
    # text를 이용해서 문자만 출력한다.
    if len(tds) > 0:
        print('tds[0].text(지점): ', tds[0].text)
        print('tds[5].text(온도): ', tds[5].text)
        print('tds[10].text(습도): ', tds[10].text)
    print('=' *  100)
```

```python
# a태그 확인하기
for tr in table.find_all('tr'):
    tds = tr.find_all('td')
    for td in tds:
        print('td = ', td)
        # td에 a태그가 있는지 확인한다. 
        print("td.find('a') = ", td.find('a'))
        print('=' * 100)
    print('*' * 100)
```

```python
# 위의 결과를 통해 a태그가 없으면, None으로 출력됨을 확인할 수 있다.
# None, 0, Null == False이다. 
# None, 0, Null이 아니면 == True
for tr in table.find_all('tr'):
    tds = tr.find_all('td')
    for td in tds:
        print('td = ', td)
        print("td.find('a') = ", td.find('a'))
        if td.find('a'):
            print("조건 td.find('a') 가 True 입니다.")
        else:
            print("조건 td.find('a') 가 False 입니다.")
        print('=' * 100)
    print('*' * 100)
```

```python
# True를 이용해서 내용 출력하기
for tr in table.find_all('tr'):
    tds = tr.find_all('td')
    for td in tds:
        if td.find('a'):
            print("td.find('a').text =", td.find('a').text)
            print("tds[5].text =", tds[5].text)
            print("tds[10].text =", tds[10].text)
    print('=' * 50)
    print('*' * 100)
```

```python
for tr in table.find_all('tr'):
    tds = tr.find_all('td')
    for td in tds:
        if td.find('a'):
            print("td.find('a').text =", td.find('a').text)
            print("tds[5].text =", tds[5].text)
            print("tds[10].text =", tds[10].text)
            data.append([tds[0].text,tds[5].text, tds[10].text])
    print(data)
    print('=' * 50)
    print('*' * 100)
```

```python
for item in data:
    print('item = ', item)
    # 리스트에서 데이터 하나씩 꺼내서 문자열 하나 만들어서 리턴
    # "문자열" 데이터와 데이터 하나에 들어간다.
    # "문자열" join(리스트) : 리스트 데이터 꺼내서 문자열 1개로 만들어줌
    row = " ^^ ".join(item)
    print('row = ', row)
    row1 = ",".join(item)
    print('row1 = ', row1)
    print('=' * 100)
```

```python
with open('weather.csv', 'w') as file:
    print('파일 저장')
    # 파일로 저장할려면 file.write를 사용한다.
    # 파일이 열어지면 읽기전용이 되니 파일을 닫았는지 확인하고 진행해야한다.
    file.write('point, temp, hum \n')
    for item in data:
        print('item = ', item)
        # ,로 나누어서 출력된다.
        row1 = ",".join(item)
        print('row1 = ', row1)
        # 파일에 데이터가 입력된다.
        # 입력하고 줄바꿈이 필요하니 \n 입력해주기
        file.write(row1+'\n')
```

```python
# 데이터 불러오기
# %matplotlib inline을 설정하면 matplotlib.pyplot의 show()함수가 없어도 그래프가 표시된다.
%matplotlib inline
# 데이터를 저장하고 처리한다.
import pandas as pd
# 그래프를 그린다.
import matplotlib as mpl
# 그래프를 그린다.
import matplotlib.pyplot as plt

# csv파일을 읽어서 Data Frame객체로 만든다.
# 현재 우리가 저장한 파일은 인코딩이 euc-kr로 되어있기 읽어올때도 euc-kr로 읽어주어야한다.
df = pd.read_csv('weather.csv', index_col = 'point', encoding = 'euc-kr')
df
```

```python
# 특별시 광역시만 모으기
# DataFrame의 loc를 이용하면 특정 인덱스의 데이터를 가져올 수 있다.
city_df = df.loc[['서울', '인천', '대전', '대구', '광주', '부산', '울산']]
city_df
```

```python
# 특별시, 광역시만 그래프 그리기
# 한글폰트가 안되서 한글 폰트를 설정해주어야한다.
font_name = mpl.font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family = font_name)

# 차트 종류, 제목, 차트 크키, 범례, 폰트 크기 설정
ax = city_df.plot(kind = 'bar', title = '날씨', figsize = (12, 4), legend = True, fontsize = 12)
# x축 정보 표시
ax.set_xlabel('도시', fontsize = 12)
# y축 정보 표시
ax.set_ylabel('기온/습도',fontsize = 12)
# 범례 지정
ax.legend(['기온', '습도'], fontsize = 12)
```
